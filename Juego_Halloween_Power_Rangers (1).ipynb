```python
import random

# Clase para representar cada personaje
class Personaje:
    def __init__(self, nombre, es_malo, vida, ataque):
        self.nombre = nombre
        self.es_malo = es_malo
        self.vida = vida
        self.ataque = ataque

    def obtener_info(self):
        return {"nombre": self.nombre, "es_malo": self.es_malo, "vida": self.vida, "ataque": self.ataque}

    def atacar(self, objetivo):
        # El personaje ataca a otro y reduce su vida
        danio = random.randint(1, self.ataque)
        print(f"{self.nombre} ataca a {objetivo.nombre} y causa {danio} puntos de daño.")
        objetivo.vida -= danio

    def esta_vivo(self):
        return self.vida > 0


# Clase para el juego
class Juego:
    def __init__(self):
        # Crear 10 personajes inspirados en mitología latinoamericana
        self.personajes = [
            Personaje("La Llorona", False, 100, 20),
            Personaje("El Chupacabras", False, 90, 15),
            Personaje("El Nahual", False, 110, 25),
            Personaje("El Hombre Caimán", False, 95, 18),
            Personaje("El Sombrerón", False, 80, 20),
            Personaje("La Cegua", False, 100, 22),
            Personaje("El Duende", False, 85, 17),
            Personaje("La Patasola", False, 100, 20),
            Personaje("El Cadejo", False, 120, 30),
            Personaje("El Familiar", True, 150, 35)  # Este es el personaje que intentará matarte
        ]
        self.personajes_seleccionados = []

    def iniciar_juego(self):
        # Seleccionar 6 personajes aleatoriamente
        self.personajes_seleccionados = random.sample(self.personajes, 6)
        print("¡Bienvenido al juego de Halloween!")
        print("Identifica al monstruo que no tiene permiso de ingresar y lucha contra él.")

        # Mostrar los personajes seleccionados
        for i, personaje in enumerate(self.personajes_seleccionados):
            print(f"{i + 1}. {personaje.nombre}")

        # Preguntar al jugador cuál es el monstruo que cree que es peligroso
        eleccion = int(input("Elige el número del monstruo sospechoso para luchar: ")) - 1
        self.luchar_con_monstruo(eleccion)

    def luchar_con_monstruo(self, eleccion):
        personaje_jugador = self.personajes_seleccionados[eleccion]
        monstruo_malo = next((p for p in self.personajes_seleccionados if p.es_malo), None)

        if monstruo_malo:
            print(f"\nTe enfrentas a {monstruo_malo.nombre}. ¡Comienza el combate!")
            while personaje_jugador.esta_vivo() and monstruo_malo.esta_vivo():
                print(f"\n{personaje_jugador.nombre} - Vida: {personaje_jugador.vida}")
                print(f"{monstruo_malo.nombre} - Vida: {monstruo_malo.vida}")
                accion = input("¿Qué quieres hacer? (1: Atacar, 2: Defender): ")

                if accion == "1":
                    personaje_jugador.atacar(monstruo_malo)
                elif accion == "2":
                    print(f"{personaje_jugador.nombre} se defiende y reduce el daño del próximo ataque.")
                    # Reducir el daño del próximo ataque del monstruo a la mitad
                    monstruo_malo.atacar(personaje_jugador)
                    personaje_jugador.vida += monstruo_malo.ataque // 2
                else:
                    print("Opción no válida. El monstruo te ataca mientras dudas.")
                    monstruo_malo.atacar(personaje_jugador)

                if monstruo_malo.esta_vivo():
                    monstruo_malo.atacar(personaje_jugador)

            if personaje_jugador.esta_vivo():
                print(f"¡Has derrotado a {monstruo_malo.nombre}! ¡Has ganado!")
            else:
                print(f"¡Has sido derrotado por {monstruo_malo.nombre}! Fin del juego.")
        else:
            print("No se encontró ningún monstruo malo en los personajes seleccionados.")

# Ejecución del juego
juego = Juego()
juego.iniciar_juego()
```
